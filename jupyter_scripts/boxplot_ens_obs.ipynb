```python
d = pd.read_csv('../tcc/data/ecc1_.csv')
d[d.vs_med=='np.nan'] = np.nan

c1_d = d[(d.data>'2018-05-23') & (d.data<'2018-05-25')]
c2_d = d[(d.data>'2018-09-23') & (d.data<'2018-09-25')]
c1_d.index = pd.to_datetime(c1_d['data'])
c2_d.index = pd.to_datetime(c2_d['data'])

c1_vento_obs = c1_d.vs_med.astype('float16').to_list()
c2_vento_obs = c2_d.vs_med.interpolate().astype('float16').to_list()
time1 = pd.to_datetime(c1_d.data)
time2 = pd.to_datetime(c2_d.data)

c1_predict = pd.read_csv('./data_ensemble.c1.csv')
c1_predict_1 = c1_predict[c1_predict['membro']==1].wspd108.astype('float16').values[37:]
c1_predict_2 = c1_predict[c1_predict['membro']==2].wspd108.astype('float16').values[37:]
c1_predict_3 = c1_predict[c1_predict['membro']==3].wspd108.astype('float16').values[37:]
c1_predict_4 = c1_predict[c1_predict['membro']==4].wspd108.astype('float16').values[37:]
c1_predict_5 = c1_predict[c1_predict['membro']==5].wspd108.astype('float16').values[37:]
c1_predict_6 = c1_predict[c1_predict['membro']==6].wspd108.astype('float16').values[37:]
c1_predict_7 = c1_predict[c1_predict['membro']==7].wspd108.astype('float16').values[37:]
c1_predict_8 = c1_predict[c1_predict['membro']==8].wspd108.astype('float16').values[37:]
c1_predict_9 = c1_predict[c1_predict['membro']==9].wspd108.astype('float16').values[37:]
c1_ens_mean = (c1_predict_1+c1_predict_2+c1_predict_3+c1_predict_4+c1_predict_5+c1_predict_6+c1_predict_7+c1_predict_8+c1_predict_9)/9
fig = plt.figure(figsize=(20,10),dpi=300)
fig.autofmt_xdate()

plt.plot(time1, c1_vento_obs)
plt.plot(time1, c1_ens_mean)
```
