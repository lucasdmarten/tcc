```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
d = pd.read_csv('../tcc/ecc1_.csv')
dCaso1 = d[(d.data>'2018-09-23') & (d.data<'2018-09-25')].interpolate()
```

```python
di = datetime.datetime(2018,9,23,0)
df = datetime.datetime(2018,9,25,0)
dates = []
while di < df:
    dates.append(di)
    di += datetime.timedelta(hours=1)
```

```python
def converttoobj(time_str):
    time_obj = datetime.datetime.strptime(time_str, '%Y-%m-%d %H:%M:%S')
    return time_obj
```

```python
box=[]
datas=[]
i=0
while i < 288:
    box.append(list(map(float,dCaso1.iloc[i:i+6,6].to_list())))
    datas.append(list(map(str,dCaso1.iloc[i:i+6,0].to_list()))[-1])
    i+=6
#datas = [converttoobj(time_str) for time_str in datas]
```

```python
dCaso1.interpolate()
```

```python
media_horaria=[]
for i in range(len(box)):
    media_horaria.append(np.mean(box[i]))
print(
        'Variancia do período: {} \n'.format(np.var(media_horaria))   
)
print(
        'Desvio padrão do período: {} \n'.format(np.std(media_horaria))
)
print(
        'Média do período: {} \n'.format(np.mean(media_horaria))
)
```

```python
newdates=[]
for i in range(len(datas)):
    date_obj = datetime.datetime.strptime(datas[i], '%Y-%m-%d %H:%M:%S')
    newdates.append('{:02d}/{:02d} - {:02d}hrs'.format(date_obj.day,date_obj.month,date_obj.hour))
```

```python
fig = plt.figure(figsize =(20, 14))
# Creating plot
plt.boxplot(x=box,vert=True,  # vertical box alignment
             patch_artist=True,  # fill with color
             labels=newdates)  # will be used to label x-ticks

degrees = 70
plt.xticks(rotation=degrees)
plt.ylabel("Velocidade (m/s)", fontsize=15)
plt.xlabel("Tempo", fontsize=15)
plt.grid(True)
plt.title("Velocidade média do vento torre ECC1",  fontsize=15)
# show plot
plt.show()
```
