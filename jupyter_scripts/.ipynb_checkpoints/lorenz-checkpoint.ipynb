```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
plt.style.use(['dark_background'])


def lorenz_parameters():
    beta = 8.0 / 3.0
    rho = 28.0
    sigma = 10.0
    t0 = 0.0
    y0 = np.array ( [ 8.0, 1.0, 1.0 ] )
    return beta, rho, sigma, t0, y0

def lorenz_deriv(t, xyz):
    beta, rho, sigma, t0, y0 = lorenz_parameters()
    dxdt = np.zeros ( 3 )
    dxdt[0] = sigma * ( xyz[1] - xyz[0] )
    dxdt[1] = xyz[0] * ( rho - xyz[2] ) - xyz[1]
    dxdt[2] = xyz[0] * xyz[1] - beta * xyz[2]
    return dxdt

beta, rho, sigma, t0, xyz0 = lorenz_parameters( )
tspan = np.array ( [ t0, 40.0 ] )
sol = solve_ivp(lorenz_deriv, tspan, xyz0)
t = sol.t
x = sol.y[0,:]
y = sol.y[1,:]
z = sol.y[2,:]


fig = plt.figure(figsize=(12,12), dpi=300)
ax = fig.gca(projection = '3d')
ax.plot (x, y, z, linewidth = 1, color = 'b')
ax.grid ( True )
ax.set_xlabel('<--- X(T) --->')
ax.set_ylabel('<--- Y(T) --->')
ax.set_zlabel('<--- Z(T) --->')
ax.set_title('Lorenz 3D Plot')
plt.savefig('lorenz_ode_3d.png')
plt.show(block = False)
plt.clf()
```
