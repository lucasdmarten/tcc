```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import pandas as pd
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.crs as ccrs
from cartopy.feature import NaturalEarthFeature
import cartopy.feature as cf
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
d = xr.open_dataset('./adaptor.mars.internal-1618413811.8133633-29306-17-03ad3c30-07e8-4f8e-ae15-6d2884bc1674.nc')
d = d.sel(longitude=-55.5,latitude=-31)
d = d.sel(time=slice('2018-09-23 00:00:00','2018-09-24 23:50:00'))
blh = d['blh']
ecc1=pd.read_csv('/media/lucasdmarten/HD/backup/projetos/tcc/tcc/data/ecc1_.csv')
ecc1[ecc1=='np.nan']=np.nan
ecc1 = ecc1[(ecc1.data>'2018-09-23') & (ecc1.data<'2018-09-25')]
ecc1['vs_med'] = ecc1['vs_med'].interpolate().astype('float')
ecc1 = ecc1.dropna()
v100 = d.variables['v100']
u100 = d.variables['u100']
speed = np.sqrt((v100**2)+(u100**2))
box=[]
datas=[]
i=0
while i < 288:
    box.append([ecc1.iloc[i:i+18,0].values[-1],np.mean(ecc1.iloc[i:i+18,6].values)])
    i+=18
p_box=[]
i=0
while i < 48:
    p_box.append([np.mean(speed[i:i+3].values),np.mean(blh[i:i+3].values)])
    i+=3
y = np.array(box)[:,1].astype('float16')
x = np.array(box)[:,0]
plt.figure(figsize=(20,10),dpi=300)
plt.plot(np.array(p_box)[:,0])
plt.plot(x,y,label='obs')
plt.xticks(rotation=40);
plt.grid(True)
plt.legend()
```

```python
print("MAE , REMQ , BIAS")
mean_absolute_error(p_box, y),mean_squared_error(p_box, y),(np.mean(p_box)-np.mean(y))
```

```python
plt.figure(figsize=(20,10),dpi=300)

plt.bar(x,np.array(p_box)[:,1],label='height CLP')
plt.xticks(rotation=40);
plt.grid(True)
plt.legend()
```
