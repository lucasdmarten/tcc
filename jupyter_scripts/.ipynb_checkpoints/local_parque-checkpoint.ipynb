```python
coords=pd.read_csv('/media/lucasdmarten/HD/backup/projetos/tcc/tcc/data/lat_lon_towers.txt')
lat_min,lat_max=coords.lat.min(),coords.lat.max()
lon_min,lon_max=coords.lon.min(),coords.lon.max()
```

```python
width = 800
height = 800
dpi = 96
resolution = '10m'

fig = plt.figure(figsize=(width / dpi, height / dpi), dpi=dpi)
ax = plt.axes(projection=ccrs.PlateCarree())
feature2=cfe.NaturalEarthFeature(category='cultural', scale='10m', 
                                name='admin_0_boundary_lines_land', 
                                color='black' ,edgecolor='face',
                                        facecolor=cfe.COLORS['land'])
ax.coastlines()
ax.add_feature(feature2)


ax.set_extent([-60,-45,-25,-36])
# ax.plot(-51.217,-30.034,'^',label='Porto Alegre',markersize=10)
ax.plot(-55.4,-30.75,'.',label='Estação A804',markersize=6, color='black')
rect = patches.Rectangle((lon_min, lat_min), .1, .1, linewidth=1, edgecolor='black', facecolor='none')
ax.add_patch(rect)
ax.plot(lon_max,lat_max,'.',label='Complexo eólico',markersize=6,color='grey')
ax.legend()
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,linestyle='--',
                  linewidth=.5, color='black', alpha=0.5)
gl.xlabels_top = True
gl.ylabels_left = True
gl.xlines = True
# gl.xlocator = mticker.FixedLocator([-70,-65,-60,-55.760,-50,-45])
# gl.ylocator = mticker.FixedLocator([-40,-38,-36,-34,-30.865,-28,-26])
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.ylabel_style = {'size': 10, 'color': 'black'}
gl.ylabel_style = {'size':8,'color': 'black', 'weight': 'bold'}
gl.xlabel_style = {'size': 10, 'color': 'gray'}
gl.xlabel_style = {'size':8,'color': 'black', 'weight': 'bold'}
# plt.title('Localização do parque',loc='left',fontsize=10)
# plt.title('Descrição',loc='right',fontsize=10)
plt.savefig('contextogeograficoparque.png')
plt.show()
```
