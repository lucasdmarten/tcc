```python
import pandas as pd
import numpy as np
```

```python
obs = pd.read_csv('/media/lucasdmarten/HD/backup/projetos/tcc/tcc/data/ecc1_.csv')
obs=obs[(obs.data>='2018-05-23')&(obs.data<'2018-05-25')]
```

```python
conjunto = {}
for caso in range(1,3):
    for membro in range(1,10):
        df = pd.read_csv('/media/lucasdmarten/HD/DADOS_TCC/RESULTADOS/CASO{}/MEMBRO{}/c{}.m{}.csv'.format(caso,membro,caso,membro))
        if caso == 1:
            speed = df.wspd108
            direction = df.wdir106
            conjunto.update({"membro{}".format(membro):{
                "wspd108":speed, "wdir106":direction
            }})
```

```python
d = pd.DataFrame(conjunto)
```

```python
soma=0
for i in range(len(d.loc['wspd108'])):
    soma += d.loc['wspd108'][i]
meanensemble = soma/len(d.loc['wspd108'])
```

```python
import matplotlib.pyplot as plt

plt.plot(meanensemble[36:-1])
obs.vs_med[obs.vs_med=='np.nan'] = np.nan
obs_vi=obs.vs_med.astype('float')
plt.plot(list(obs_vi))
```
