```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "/media/lucasdmarten/HD/DADOS_TCC/dados_estacao/historicos/dados_A804_M_2001-11-21_2020-01-01.csv"
d = pd.read_csv(path,skiprows=9,delimiter=';').interpolate()
d.index = pd.to_datetime(d["Data Medicao"])
```

```python
d.columns
```

```python
df = d.groupby(d.index.month).mean()
df.index = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
mean_wind_month=df["TEMPERATURA MEDIA, MENSAL (AUT)(°C)"].tolist()
```

```python
np.linspace(10,30,20)
```

```python
fig = plt.figure(figsize =(20, 14))
# Creating plot
plt.plot(df.index,mean_wind_month)

plt.ylabel("Temperatura em °C", fontsize=25)
plt.yticks(fontsize=15)
plt.ylim(10,25)
plt.xlabel("Mês", fontsize=25)
plt.xticks(fontsize=15)

plt.grid(True)
plt.title("Temperatura média mensal",  fontsize=25, loc='left')
plt.title("Período: 1990 à 2020\nSantana do Livramento - A804",  fontsize=15, loc='right')

plt.savefig('month_temperature_mean_A804.png',dpi=300)
plt.show()
```
